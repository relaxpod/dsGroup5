```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
url = "https://raw.githubusercontent.com/relaxpod/dsGroup5/refs/heads/main/states_insurance_dataset_regions.csv"
df = pd.read_csv(url)
df
```

```python
# Bar Plot for Uninsured Rate
plt.figure(figsize=(8, 5))
plt.barh(df["US Regions"], df["Uninsured Rate (2015) Average"], color="skyblue")
plt.xlabel("Uninsured Rate")
plt.title("The Average Uninsured Rate for Each US Region")
plt.show()
```

```python
url2 = "https://raw.githubusercontent.com/relaxpod/dsGroup5/refs/heads/main/states_insurance_dataset_temp.csv"
df2 = pd.read_csv(url2)
df2
```

```python
# More accurate heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df2.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Heatmap of Correlations")
plt.show()
```
